```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# ----------------------------
# LOAD DATA
# ----------------------------
df = pd.read_csv("/hr_analytics_data.csv")
```

```python
# date conversion
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"], errors="coerce")
df["Last_Promotion_Date"] = pd.to_datetime(df["Last_Promotion_Date"], errors="coerce")

sns.set(style="whitegrid")
```

```python
# KPI INSIGHTS
# ----------------------------
print("Total Employees:", df["Employee_ID"].nunique())
print("Average Salary:", round(df["Salary"].mean(), 2))
print("Average Experience:", round(df["Experience_Years"].mean(), 2))
print("Attrition Count:", df["Attrition"].sum())
print("Average Performance Rating:", round(df["Performance_Rating"].mean(), 2))
```

```python
# ----------------------------
# 1. DEPARTMENT DISTRIBUTION
# ----------------------------
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Department")
plt.title("Employees by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# ----------------------------
# 2. JOB ROLE DISTRIBUTION
# ----------------------------
plt.figure(figsize=(12,5))
role_counts = df["Job_Role"].value_counts().head(10)
sns.barplot(x=role_counts.index, y=role_counts.values)
plt.title("Top Job Roles")
plt.xticks(rotation=45)
plt.show()
```

```python
# ----------------------------
# 3. GENDER DISTRIBUTION
# ----------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution")
plt.show()
```

```python
# ----------------------------
# 4. AGE DISTRIBUTION
# ----------------------------
plt.figure(figsize=(8,4))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# ----------------------------
# 5. SALARY BY DEPARTMENT
# ----------------------------
plt.figure(figsize=(12,5))
sns.boxplot(data=df, x="Department", y="Salary")
plt.title("Salary Distribution by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# ----------------------------
# 6. PERFORMANCE RATING
# ----------------------------
plt.figure(figsize=(7,4))
sns.countplot(data=df, x="Performance_Rating")
plt.title("Performance Rating Distribution")
plt.show()
```

```python
# ----------------------------
# 7. ATTRITION ANALYSIS
# ----------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Attrition")
plt.title("Attrition Distribution")
plt.show()
```

```python
# ----------------------------
# 8. ATTRITION BY DEPARTMENT
# ----------------------------
plt.figure(figsize=(12,5))
sns.countplot(data=df, x="Department", hue="Attrition")
plt.title("Attrition by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# ----------------------------
# 9. EXPERIENCE VS SALARY
# ----------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="Experience_Years",
    y="Salary",
    hue="Department"
)
plt.title("Experience vs Salary")
plt.show()
```

```python
# ----------------------------
# 10. OVERTIME VS ATTRITION
# ----------------------------
plt.figure(figsize=(7,4))
sns.countplot(data=df, x="Overtime", hue="Attrition")
plt.title("Overtime vs Attrition")
plt.show()
```

```python
# ----------------------------
# 11. TRAINING HOURS VS PERFORMANCE
# ----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="Performance_Rating",
    y="Training_Hours"
)
plt.title("Training Hours by Performance Rating")
plt.show()
```

```python
# ----------------------------
# 12. PROJECTS HANDLED
# ----------------------------
plt.figure(figsize=(8,4))
sns.histplot(df["Projects_Handled"], bins=15, kde=True)
plt.title("Projects Handled Distribution")
plt.show()
```

```python

# ----------------------------
# 14. JOINING TREND (PLOTLY)
# ----------------------------
joining_trend = (
    df.groupby(df["Joining_Date"].dt.to_period("M"))
    .size()
    .reset_index(name="Employees")
)

joining_trend["Joining_Date"] = joining_trend["Joining_Date"].astype(str)

fig = px.line(
    joining_trend,
    x="Joining_Date",
    y="Employees",
    title="Employee Joining Trend"
)
fig.show()
```

```python
# ----------------------------
# 15. INTERACTIVE ATTRITION BY JOB ROLE
# ----------------------------
attrition_role = (
    df.groupby("Job_Role")["Attrition"]
    .sum()
    .reset_index()
    .sort_values("Attrition", ascending=False)
)

fig2 = px.bar(
    attrition_role,
    x="Job_Role",
    y="Attrition",
    title="Attrition by Job Role"
)
fig2.show()
```
