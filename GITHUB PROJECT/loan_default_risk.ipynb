```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load file
df = pd.read_csv("C:/Users/hp/OneDrive/Desktop/Project github/loan_default_risk_data.csv")
```

```python
# Date format
df["Application_Date"] = pd.to_datetime(df["Application_Date"])
```

```python
# -----------------------------
# KPI VALUES
# -----------------------------
total_loans = df["Loan_ID"].nunique()
total_amount = df["Loan_Amount"].sum()
avg_interest = df["Interest_Rate"].mean()
default_rate = df["Default_Flag"].mean() * 100
avg_credit = df["Credit_Score"].mean()

print("Total Loans:", total_loans)
print("Total Loan Amount:", total_amount)
print("Average Interest Rate:", round(avg_interest, 2))
print("Default Rate %:", round(default_rate, 2))
print("Average Credit Score:", round(avg_credit, 2))
```

```python
# -----------------------------
# 1. Loan Applications Trend
# -----------------------------
trend = df.groupby(df["Application_Date"].dt.to_period("M")).size().reset_index(name="Applications")
trend["Application_Date"] = trend["Application_Date"].astype(str)

fig1 = px.line(
    trend,
    x="Application_Date",
    y="Applications",
    title="Loan Applications Over Time"
)
fig1.show()
```

```python
# -----------------------------
# 2. Loan Amount by City
# -----------------------------
city_loan = df.groupby("City")["Loan_Amount"].sum().reset_index()

fig2 = px.bar(
    city_loan.sort_values("Loan_Amount", ascending=False).head(10),
    x="City",
    y="Loan_Amount",
    title="Top 10 Cities by Loan Amount"
)
fig2.show()
```

```python
# -----------------------------
# 3. Loan Purpose Distribution
# -----------------------------
purpose = df["Loan_Purpose"].value_counts().reset_index()
purpose.columns = ["Loan_Purpose", "Count"]

fig3 = px.pie(
    purpose,
    names="Loan_Purpose",
    values="Count",
    title="Loan Purpose Distribution"
)
fig3.show()
```

```python
# -----------------------------
# 4. Default Rate by Employment Type
# -----------------------------
default_emp = df.groupby("Employment_Type")["Default_Flag"].mean().reset_index()
default_emp["Default_Flag"] *= 100

fig4 = px.bar(
    default_emp,
    x="Employment_Type",
    y="Default_Flag",
    title="Default Rate by Employment Type"
)
fig4.show()
```

```python
# -----------------------------
# 5. Credit Score vs Loan Amount
# -----------------------------
fig5 = px.scatter(
    df,
    x="Credit_Score",
    y="Loan_Amount",
    color="Default_Flag",
    title="Credit Score vs Loan Amount"
)
fig5.show()
```

```python
# -----------------------------
# 6. Age Group Analysis
# -----------------------------
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[18, 25, 35, 45, 55, 65],
    labels=["18-25", "26-35", "36-45", "46-55", "56-65"]
)

age_loan = df.groupby("Age_Group")["Loan_Amount"].sum().reset_index()

fig6 = px.bar(
    age_loan,
    x="Age_Group",
    y="Loan_Amount",
    title="Loan Amount by Age Group"
)
fig6.show()
```

```python
# -----------------------------
# 7. Income vs EMI
# -----------------------------
fig7 = px.scatter(
    df,
    x="Income",
    y="EMI",
    color="Default_Flag",
    title="Income vs EMI"
)
fig7.show()
```

```python
# -----------------------------
# 8. Debt-to-Income by Default
# -----------------------------
fig8 = px.box(
    df,
    x="Default_Flag",
    y="Debt_to_Income",
    title="Debt-to-Income Ratio by Default"
)
fig8.show()
```

```python
# -----------------------------
# 9. Payment History Distribution
# -----------------------------
fig9 = px.histogram(
    df,
    x="Payment_History",
    nbins=20,
    title="Payment History Distribution"
)
fig9.show()
```

```python
# -----------------------------
# 10. Correlation Heatmap
# -----------------------------
num_cols = [
    "Age", "Income", "Credit_Score", "Loan_Amount",
    "Loan_Term_Months", "Interest_Rate", "EMI",
    "Existing_Loans", "Debt_to_Income", "Payment_History"
]

corr = df[num_cols].corr()

fig10 = px.imshow(
    corr,
    text_auto=True,
    title="Correlation Heatmap"
)
fig10.show()
```
