```python
import pandas as pd
from dash import Dash, dcc, html
import plotly.express as px
```

```python
# Load Data
df = pd.read_excel("C:/Users/hp/OneDrive/Desktop/Project github/customer_segmentation_purchase_behavior.xlsx")
```

```python
# Convert Date
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
```

```python
# KPIs
total_revenue = df['Purchase_Amount'].sum()
total_customers = df['Customer_ID'].nunique()
avg_purchase = df['Purchase_Amount'].mean()
avg_clv = df['Customer_Lifetime_Value'].mean()
```

```python
# Charts

# Category Sales
category_sales = df.groupby(
    'Product_Category'
)['Purchase_Amount'].sum().reset_index()

fig_category = px.bar(
    category_sales,
    x='Product_Category',
    y='Purchase_Amount',
    title='Revenue by Product Category'
)
```

```python
# Gender Analysis
gender_sales = df.groupby(
    'Gender'
)['Purchase_Amount'].sum().reset_index()

fig_gender = px.pie(
    gender_sales,
    names='Gender',
    values='Purchase_Amount',
    hole=0.4,
    title='Gender-wise Purchases'
)
```

```python
# City Sales
city_sales = df.groupby(
    'City'
)['Purchase_Amount'].sum().reset_index()

fig_city = px.bar(
    city_sales,
    x='Purchase_Amount',
    y='City',
    orientation='h',
    title='City-wise Revenue'
)
```

```python
# Income vs Purchase
fig_income = px.scatter(
    df,
    x='Income',
    y='Purchase_Amount',
    color='Gender',
    title='Income vs Purchase Amount'
)
```

```python
# CLV Distribution
fig_clv = px.histogram(
    df,
    x='Customer_Lifetime_Value',
    nbins=30,
    title='Customer Lifetime Value Distribution'
)
```

```python
# Monthly Sales
monthly_sales = df.groupby(
    df['Purchase_Date'].dt.month_name()
)['Purchase_Amount'].sum().reset_index()

fig_monthly = px.line(
    monthly_sales,
    x='Purchase_Date',
    y='Purchase_Amount',
    markers=True,
    title='Monthly Sales Trend'
)
```
