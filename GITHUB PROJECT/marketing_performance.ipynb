Transformation

```python
# Revenue and spend by channel
channel_summary = df.groupby("Channel")[["Revenue", "Spend"]].sum()
print(channel_summary)
```

```python
#Campaign count by region
region_count = df.groupby("Region")["Campaign_ID"].count()
print(region_count)
```

```python
#Average CTR by device
device_ctr = df.groupby("Device_Type")["CTR"].mean()
print(device_ctr)
```

```python
#Revenue summary by channel
channel_agg = df.groupby("Channel").agg({
    "Revenue": ["sum", "mean", "max"],
    "Spend": ["sum", "mean"],
    "Clicks": ["sum"],
    "Conversions": ["sum"]
})

print(channel_agg)
```

```python
#ROAS summary by country
country_agg = df.groupby("Country").agg({
    "ROAS": ["mean", "max"],
    "Revenue": "sum",
    "Spend": "sum"
})

print(country_agg)
```

```python
#Pivot Table
pivot_rev = pd.pivot_table(
    df,
    values="Revenue",
    index="Channel",
    columns="Region",
    aggfunc="sum",
    fill_value=0
)

print(pivot_rev)
```

```python
#Conversions by device and gender
pivot_conv = pd.pivot_table(
    df,
    values="Conversions",
    index="Device_Type",
    columns="Audience_Gender",
    aggfunc="sum",
    fill_value=0
)

print(pivot_conv)
```

```python
#Channel total spend
channel_spend = df.groupby("Channel")["Spend"].sum().reset_index()
channel_spend.rename(columns={"Spend": "Channel_Total_Spend"}, inplace=True)
```

```python
#Merge back into original dataframe
df = pd.merge(df, channel_spend, on="Channel", how="left")
print(df.head())
```

```python
#Best performing channels
best_channels = df.groupby("Channel").agg({
    "Revenue": "sum",
    "Spend": "sum",
    "Conversions": "sum"
})

best_channels["ROAS"] = best_channels["Revenue"] / best_channels["Spend"]

print(best_channels.sort_values("ROAS", ascending=False))
```

```python
#Create calculated metrics
df["Profit"] = df["Revenue"] - df["Spend"]
df["Conversion_Per_Click"] = df["Conversions"] / df["Clicks"]
df["Revenue_Per_Click"] = df["Revenue"] / df["Clicks"]
df["Cost_Per_Lead"] = df["Spend"] / df["Leads"]
```

Visualization Dashboards

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/marketing_performance_70000_rows.csv")
```

```python
total_revenue = df["Revenue"].sum()
total_spend = df["Spend"].sum()
total_clicks = df["Clicks"].sum()
total_conversions = df["Conversions"].sum()
avg_roas = df["ROAS"].mean()
```

```python
#Revenue by Channel
channel_rev = df.groupby("Channel")["Revenue"].sum().sort_values()

plt.figure(figsize=(8,5))
channel_rev.plot(kind="barh")
plt.title("Revenue by Channel")
plt.xlabel("Revenue")
plt.show()
```

```python
#Spend vs Revenue (comparison)
channel_perf = df.groupby("Channel")[["Spend", "Revenue"]].sum()

plt.figure(figsize=(8,5))
channel_perf.plot(kind="bar")
plt.title("Spend vs Revenue by Channel")
plt.ylabel("Amount")
plt.show()
```

```python
#Daily campaign trend
df["Start_Date"] = pd.to_datetime(df["Start_Date"])

daily_rev = df.groupby("Start_Date")["Revenue"].sum()

plt.figure(figsize=(12,5))
daily_rev.plot()
plt.title("Revenue Trend Over Time")
plt.ylabel("Revenue")
plt.show()
```

```python
#CTR by Device Type
device_ctr = df.groupby("Device_Type")["CTR"].mean()

plt.figure(figsize=(7,4))
device_ctr.plot(kind="bar")
plt.title("Average CTR by Device")
plt.ylabel("CTR")
plt.show()
```

```python
#Conversions by Region
region_conv = df.groupby("Region")["Conversions"].sum().sort_values()

plt.figure(figsize=(8,5))
region_conv.plot(kind="barh")
plt.title("Conversions by Region")
plt.xlabel("Conversions")
plt.show()
```

```python
#ROAS by Channel
channel_roas = df.groupby("Channel")["ROAS"].mean().sort_values()

plt.figure(figsize=(8,5))
channel_roas.plot(kind="barh")
plt.title("Average ROAS by Channel")
plt.xlabel("ROAS")
plt.show()
```

```python
#Funnel chart idea
funnel = pd.Series({
    "Impressions": df["Impressions"].sum(),
    "Clicks": df["Clicks"].sum(),
    "Leads": df["Leads"].sum(),
    "Conversions": df["Conversions"].sum(),
    "Sales": df["Sales"].sum()
})

plt.figure(figsize=(8,5))
funnel.plot(kind="bar")
plt.title("Marketing Funnel")
plt.ylabel("Volume")
plt.show()
```
