```python
# Load Data
df = pd.read_csv("C:/Users/hp/OneDrive/Desktop/Project github/air_pollution_analysis_data.csv")
```

```python
# Convert Date
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Widgets
city_select = pn.widgets.Select(
    name='City',
    options=list(df['City'].unique()))
```

```python
# Filter Function
def get_filtered_data():
    return df[df['City'] == city_select.value]
```

```python
# KPI Function
def kpi():
    data = get_filtered_data()

    avg_aqi = round(data['AQI'].mean(), 2)
    max_aqi = round(data['AQI'].max(), 2)
    avg_health = round(data['Health_Index'].mean(), 2)

    return pn.Row(
        pn.indicators.Number(name='Average AQI', value=avg_aqi),
        pn.indicators.Number(name='Max AQI', value=max_aqi),
        pn.indicators.Number(name='Health Index', value=avg_health)
    )
```

```python
# Plotly AQI Trend
def aqi_trend():
    data = get_filtered_data()

    fig = px.line(
        data,
        x='Date',
        y='AQI',
        title='AQI Trend'
    )

    return pn.pane.Plotly(fig)
```

```python
# Plotly Scatter
def pollution_scatter():
    data = get_filtered_data()

    fig = px.scatter(
        data,
        x='PM2_5',
        y='PM10',
        color='AQI',
        size='Population_Density',
        title='PM2.5 vs PM10'
    )
    return pn.pane.Plotly(fig)
```

```python
# Seaborn Heatmap
def correlation_heatmap():
    data = get_filtered_data()

    corr = data.select_dtypes(include='number').corr()

    fig, ax = plt.subplots(figsize=(10,6))

    sns.heatmap(
        corr,
        annot=True,
        cmap='coolwarm',
        ax=ax
    )

    plt.title("Correlation Heatmap")

    return pn.pane.Matplotlib(fig, tight=True)
```

```python
# Temperature vs AQI
def temp_vs_aqi():
    data = get_filtered_data()

    fig = px.scatter(
        data,
        x='Temperature_C',
        y='AQI',
        color='Traffic_Level',
        title='Temperature vs AQI'
    )

    return pn.pane.Plotly(fig)
```

```python
pn.panel(aqi_trend())
pn.panel(temp_vs_aqi())
pn.panel(aqi_trend())
pn.panel(aqi_trend())
pn.panel(aqi_trend())
```

```python
 # Dashboard Layout
dashboard = pn.Column(
    "Air Quality Dashboard",

    city_select,

    pn.panel(kpi),

    pn.Row(
        pn.panel(aqi_trend),
        pn.panel(temp_vs_aqi)
    ),

    pn.Row(
        pn.panel(pollution_scatter),
        pn.panel(correlation_heatmap)
    )
)

```

```python

# Serve Dashboard
dashboard.servable()
```
