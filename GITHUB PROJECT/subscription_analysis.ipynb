```python
import pandas as pd

df = pd.read_csv("C:/Users/hp/OneDrive/Desktop/Project github/subscription_analysis_data.csv")

print(df.head())
print(df.info())
```

Data Cleaning

```python
# Check missing values
print(df.isnull().sum())

# Convert dates
df['Start_Date'] = pd.to_datetime(df['Start_Date'])


# Remove duplicates
df.drop_duplicates(inplace=True)

# Basic statistics
print(df.describe())
```

KPIs

```python
total_customers = df['Customer_ID'].nunique()
print("Total Customers:", total_customers)
```

```python
avg_satisfaction = df['Satisfaction_Score'].mean()

print("Average Satisfaction:", round(avg_satisfaction,2))
```

```python
churn_rate = (df['Churn_Flag'].sum() / len(df)) * 100

print("Churn Rate:", round(churn_rate,2), "%")
```

Data Analysis

```python
subscription_analysis = df.groupby('Subscription_Type').agg({
    'Customer_ID':'count',
    'Monthly_Fee':'sum',
    'Satisfaction_Score':'mean'
})

print(subscription_analysis)
```

```python
city_revenue = df.groupby('City')['Monthly_Fee'].sum().sort_values(ascending=False)

print(city_revenue)
```

```python
gender_analysis = df.groupby('Gender')['Customer_ID'].count()

print(gender_analysis)
```

```python
churn_satisfaction = df.groupby('Churn_Flag')['Satisfaction_Score'].mean()

print(churn_satisfaction)
```

```python
payment_analysis = df.groupby('Payment_Method')['Monthly_Fee'].sum()

print(payment_analysis)
```

Visualizations

```python
import matplotlib.pyplot as plt

subscription_revenue = df.groupby('Subscription_Type')['Monthly_Fee'].sum()

subscription_revenue.plot(kind='bar')

plt.title("Revenue by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Revenue")

plt.show()
```

```python
df['Churn_Flag'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Churn Distribution")

plt.show()
```

```python
city_revenue.head(10).plot(kind='bar')

plt.title("Top 10 Cities by Revenue")

plt.show()
```

```python
df['Satisfaction_Score'].hist(bins=10)

plt.title("Customer Satisfaction Distribution")

plt.xlabel("Score")

plt.show()
```

```python
import plotly.express as px

fig = px.bar(
    df,
    x='Subscription_Type',
    y='Monthly_Fee',
    color='Subscription_Type',
    title='Subscription Revenue Dashboard'
)

fig.show()
```
