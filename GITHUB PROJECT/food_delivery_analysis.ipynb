VISUALIZATION DASHBOARD

```python
import plotly.express as px
```

```python
total_orders = df['Order_ID'].nunique()
total_revenue = df['Order_Value'].sum()
avg_delivery = df['Delivery_Time_Minutes'].mean()
avg_rating = df['Rating'].mean()
```

Revenue by City

```python
city_sales = df.groupby('City')['Order_Value'].sum().reset_index()

fig = px.bar(
    city_sales,
    x='City',
    y='Order_Value',
    color='City',
    title='Revenue by City'
)
```

```python
daily_orders = df.groupby('Order_Date')['Order_ID'].count().reset_index()

fig = px.line(
    daily_orders,
    x='Order_Date',
    y='Order_ID',
    title='Daily Orders Trend'
)
```

```python
cuisine = df.groupby('Cuisine_Type')['Order_ID'].count().reset_index()

fig = px.pie(
    cuisine,
    names='Cuisine_Type',
    values='Order_ID',
    hole=0.5
)
```

```python
fig = px.scatter(
    df,
    x='Distance_km',
    y='Delivery_Time_Minutes',
    color='Traffic_Level',
    size='Order_Value'
)
```

```python
fig = px.box(
    df,
    x='Weather',
    y='Delivery_Time_Minutes',
    color='Weather'
)
```

```python
fig = px.density_heatmap(
    df,
    x='Traffic_Level',
    y='Delivery_Time_Minutes'
)
```

```python
fig = px.pie(
    df,
    names='Payment_Method',
    title='Payment Methods'
)
```

```python
fig = px.bar(
    df,
    x='Is_Peak_Hour',
    y='Order_Value',
    color='Order_Status'
)
```

```python
fig = px.histogram(
    df,
    x='Rating',
    nbins=10
)
```
