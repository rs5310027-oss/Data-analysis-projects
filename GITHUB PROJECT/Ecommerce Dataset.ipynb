Visualization

```python
df.groupby('Gender')['Amount'].sum().plot(kind='bar')

plt.title("Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Sales")
plt.show()
```

```python
df.groupby('Month')['Amount'].sum().plot(kind='line')

plt.title("Monthly Sales")
plt.show()
```

```python
df.groupby('Category')['Amount'].sum().plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")
plt.show()
```

Advanced Operations

```python
df['Sales_Tax'] = df['Amount'] * 0.18
```

```python
df['High_Value'] = np.where(df['Amount'] > 5000, 'Yes', 'No')
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Day_Name'] = df['Date'].dt.day_name()
```

```python
df.to_csv("cleaned_sales_data.csv", index=False)
```

```python
total_sales = df['Amount'].sum()

total_orders = df['Order ID'].nunique()

avg_order = df['Amount'].mean()

top_category = df.groupby('Category')['Amount'].sum().idxmax()

print("Total Sales:", total_sales)
print("Total Orders:", total_orders)
print("Average Order Value:", avg_order)
print("Top Category:", top_category)
```
