```python
total_matches = df["Match_ID"].nunique()
print("Total Matches:", total_matches)
```

```python
top_players = df.groupby("Player_Name")["Runs"].sum().sort_values(ascending=False).head(10)
print(top_players)
```

```python
team_runs = df.groupby("Team")["Runs"].sum().sort_values(ascending=False)
print(team_runs)
```

```python
df["Win_Flag"] = (df["Team"] == df["Winner"]).astype(int)

win_stats = df.groupby("Team").agg(
    Matches=("Match_ID","count"),
    Wins=("Win_Flag","sum")
)

win_stats["Win_Percentage"] = (win_stats["Wins"] / win_stats["Matches"]) * 100

print(win_stats)
```

Data Analysis and Visualizations

```python
import matplotlib.pyplot as plt

top_players.plot(kind='bar')
plt.title("Top Players by Runs")
plt.xlabel("Player")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
win_stats["Win_Percentage"].plot(kind='bar')
plt.title("Team Win Percentage")
plt.ylabel("Win %")
plt.show()
```

```python
import seaborn as sns

sns.barplot(x=top_players.values, y=top_players.index)
plt.title("Top Players (Seaborn)")
plt.show()
```

```python
import numpy as np

avg_runs = np.mean(df["Runs"])
print("Average Runs:", avg_runs)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df[["Runs","Balls","Strike_Rate","Wickets","Economy"]].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.histplot(df["Runs"], bins=30)
plt.title("Runs Distribution")
plt.show()
```

```python
top_allrounders = df.groupby("Player_Name").agg({
    "Runs":"sum",
    "Wickets":"sum"
})

top_allrounders = top_allrounders[
    (top_allrounders["Runs"] > 300) & (top_allrounders["Wickets"] > 10)
].sort_values("Runs", ascending=False)

print(top_allrounders.head(10))
```

```python
team_stats = df.groupby("Team").agg({
    "Runs":"sum",
    "Wickets":"sum",
    "Match_ID":"count"
})

team_stats["Avg_Runs"] = team_stats["Runs"] / team_stats["Match_ID"]

print(team_stats)
```

```python
sns.boxplot(x="Team", y="Runs", data=df)
plt.xticks(rotation=45)
plt.title("Team-wise Runs Distribution")
plt.show()
```
