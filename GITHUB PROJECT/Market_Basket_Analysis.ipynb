Visualization Dashboard

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load dataset
df = pd.read_csv("/market_basket_analysis_data.csv")
```

```python
# -----------------------------
# Basic preprocessing
# -----------------------------
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
# Create useful time columns
df['Month'] = df['Transaction_Date'].dt.to_period('M').astype(str)
df['Day_Name'] = df['Transaction_Date'].dt.day_name()
```

```python
# -----------------------------
# KPI values
# -----------------------------
total_sales = df['Total_Amount'].sum()
total_transactions = df['Transaction_ID'].nunique()
avg_basket = df['Basket_Size'].mean()
avg_discount = df['Discount'].mean()
loyalty_rate = df['Loyalty_Customer'].mean() * 100

print("Total Sales:", round(total_sales, 2))
print("Total Transactions:", total_transactions)
print("Average Basket Size:", round(avg_basket, 2))
print("Average Discount:", round(avg_discount, 2))
print("Loyalty Customer %:", round(loyalty_rate, 2))
```

```python
# ==========================================================
# 1. Sales Trend Over Time
# ==========================================================
sales_trend = df.groupby('Month')['Total_Amount'].sum().reset_index()

fig1 = px.line(
    sales_trend,
    x='Month',
    y='Total_Amount',
    title='Sales Trend Over Time'
)
fig1.show()
```

```python
# ==========================================================
# 2. Sales by Region
# ==========================================================
region_sales = df.groupby('Region')['Total_Amount'].sum().reset_index()

fig2 = px.bar(
    region_sales,
    x='Region',
    y='Total_Amount',
    title='Sales by Region'
)
fig2.show()
```

```python
# ==========================================================
# 3. Product Category Performance
# ==========================================================
category_sales = df.groupby('Product_Category')['Total_Amount'].sum().reset_index()

fig3 = px.bar(
    category_sales,
    x='Product_Category',
    y='Total_Amount',
    title='Sales by Product Category'
)
fig3.show()
```

```python
# ==========================================================
# 4. Top 10 Products
# ==========================================================
top_products = (
    df.groupby('Product_Name')['Total_Amount']
    .sum()
    .nlargest(10)
    .reset_index()
)

fig4 = px.bar(
    top_products,
    x='Total_Amount',
    y='Product_Name',
    orientation='h',
    title='Top 10 Products'
)
fig4.show()
```

```python
# ==========================================================
# 5. Payment Method Distribution
# ==========================================================
payment_sales = df.groupby('Payment_Method')['Total_Amount'].sum().reset_index()

fig5 = px.pie(
    payment_sales,
    names='Payment_Method',
    values='Total_Amount',
    title='Payment Method Distribution'
)
fig5.show()
```

```python
# ==========================================================
# 6. Brand Performance
# ==========================================================
brand_sales = (
    df.groupby('Brand')['Total_Amount']
    .sum()
    .nlargest(10)
    .reset_index()
)

fig6 = px.bar(
    brand_sales,
    x='Brand',
    y='Total_Amount',
    title='Top 10 Brands'
)
fig6.show()
```

```python
# ==========================================================
# 7. Weekend vs Weekday Sales
# ==========================================================
weekend_sales = df.groupby('Weekend_Flag')['Total_Amount'].sum().reset_index()

fig7 = px.bar(
    weekend_sales,
    x='Weekend_Flag',
    y='Total_Amount',
    title='Weekend vs Weekday Sales'
)
fig7.show()
```

```python
# ==========================================================
# 8. Promotion Impact
# ==========================================================
promo_sales = df.groupby('Promotion_Applied')['Total_Amount'].mean().reset_index()

fig8 = px.bar(
    promo_sales,
    x='Promotion_Applied',
    y='Total_Amount',
    title='Average Sales: Promotion Impact'
)
fig8.show()
```

```python
# ==========================================================
# 9. Purchase Time Analysis
# ==========================================================
time_sales = df.groupby('Purchase_Time')['Total_Amount'].sum().reset_index()

fig9 = px.bar(
    time_sales,
    x='Purchase_Time',
    y='Total_Amount',
    title='Sales by Purchase Time'
)
fig9.show()
```

```python
# ==========================================================
# 10. Loyalty Customer Impact
# ==========================================================
loyalty_sales = df.groupby('Loyalty_Customer')['Total_Amount'].mean().reset_index()

fig10 = px.bar(
    loyalty_sales,
    x='Loyalty_Customer',
    y='Total_Amount',
    title='Average Spend: Loyalty vs Non-Loyalty'
)
fig10.show()
```
