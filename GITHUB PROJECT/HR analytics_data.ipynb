Visualization Dashboards

```python
import pandas as pd
import plotly.express as px
```

```python
# Load data
df = pd.read_csv("C:/Users/hp/OneDrive/Desktop/Project github/hr_analytics_data.csv")
```

```python
# Date columns
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])
df["Joining_Month"] = df["Joining_Date"].dt.to_period("M").astype(str)
```

```python
# ==========================
# KPI (optional)
# ==========================
print("Total Employees:", df["Employee_ID"].nunique())
print("Average Salary:", round(df["Salary"].mean(), 2))
print("Attrition Rate:", round(df["Attrition"].mean() * 100, 2), "%")
```

```python
# ==========================
# 1. Employee count by department
# ==========================
dept = df.groupby("Department")["Employee_ID"].count().reset_index()

fig1 = px.bar(
    dept,
    x="Department",
    y="Employee_ID",
    title="Employee Count by Department"
)
fig1.show()
```

```python
# ==========================
# 2. Attrition by department
# ==========================
attrition = df.groupby("Department")["Attrition"].mean().reset_index()

fig2 = px.bar(
    attrition,
    x="Department",
    y="Attrition",
    title="Attrition Rate by Department"
)
fig2.show()
```

```python
# ==========================
# 3. Salary by job role
# ==========================
salary_role = df.groupby("Job_Role")["Salary"].mean().reset_index()

fig3 = px.bar(
    salary_role,
    x="Job_Role",
    y="Salary",
    title="Average Salary by Job Role"
)
fig3.show()
```

```python
# ==========================
# 4. Performance vs experience
# ==========================
fig4 = px.scatter(
    df,
    x="Experience_Years",
    y="Performance_Rating",
    size="Salary",
    color="Department",
    title="Performance vs Experience"
)
fig4.show()
```

```python
# ==========================
# 5. Gender distribution
# ==========================
gender = df.groupby("Gender")["Employee_ID"].count().reset_index()

fig5 = px.pie(
    gender,
    names="Gender",
    values="Employee_ID",
    title="Gender Distribution"
)
fig5.show()
```

```python
# ==========================
# 6. Overtime vs attrition
# ==========================
overtime = df.groupby("Overtime")["Attrition"].mean().reset_index()

fig6 = px.bar(
    overtime,
    x="Overtime",
    y="Attrition",
    title="Overtime vs Attrition"
)
fig6.show()
```

```python
# ==========================
# 7. Training hours vs performance
# ==========================
fig7 = px.scatter(
    df,
    x="Training_Hours",
    y="Performance_Rating",
    color="Department",
    title="Training Hours vs Performance"
)
fig7.show()
```

```python
# ==========================
# 8. Hiring trend
# ==========================
hiring = df.groupby("Joining_Month")["Employee_ID"].count().reset_index()

fig8 = px.line(
    hiring,
    x="Joining_Month",
    y="Employee_ID",
    title="Hiring Trend Over Time"
)
fig8.show()
```

```python
# ==========================
# 9. Sick leaves by department
# ==========================
sick = df.groupby("Department")["Sick_Leaves"].mean().reset_index()

fig9 = px.bar(
    sick,
    x="Department",
    y="Sick_Leaves",
    title="Average Sick Leaves by Department"
)
fig9.show()
```

```python
# ==========================
# 10. Manager rating by department
# ==========================
mgr = df.groupby("Department")["Manager_Rating"].mean().reset_index()

fig10 = px.bar(
    mgr,
    x="Department",
    y="Manager_Rating",
    title="Average Manager Rating by Department"
)
fig10.show()
```
