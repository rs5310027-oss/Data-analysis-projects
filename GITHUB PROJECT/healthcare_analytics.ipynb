```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python

# Load data
df = pd.read_csv("C:/Users/hp/OneDrive/Desktop/Project github/healthcare_analytics_data.csv")
```

```python
 # Date conversion
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], errors="coerce")
df["Discharge_Date"] = pd.to_datetime(df["Discharge_Date"], errors="coerce")

sns.set(style="whitegrid")
```

```python
# ---------------------------------------------------
# KPI INSIGHTS
# ---------------------------------------------------
print("Total Patients:", df["Patient_ID"].nunique())
print("Average Age:", round(df["Age"].mean(), 1))
print("Average Length of Stay:", round(df["Length_of_Stay"].mean(), 1))
print("Total Treatment Cost:", round(df["Treatment_Cost"].sum(), 2))
print("Average Treatment Cost:", round(df["Treatment_Cost"].mean(), 2))
```

```python
# ---------------------------------------------------
# 1. PATIENT ADMISSION TREND
# ---------------------------------------------------
admission_trend = df.groupby("Admission_Date")["Patient_ID"].count()

plt.figure(figsize=(14,5))
plt.plot(admission_trend.index, admission_trend.values)
plt.title("Patient Admissions Over Time")
plt.xlabel("Admission Date")
plt.ylabel("Admissions")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 2. GENDER DISTRIBUTION
# ---------------------------------------------------
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution")
plt.show()
```

```python
# ---------------------------------------------------
# 3. AGE DISTRIBUTION
# ---------------------------------------------------
plt.figure(figsize=(8,4))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# ---------------------------------------------------
# 4. TOP CITIES BY PATIENT COUNT
# ---------------------------------------------------
city_counts = df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=city_counts.index, y=city_counts.values)
plt.title("Top 10 Cities by Patient Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 5. DEPARTMENT WORKLOAD
# ---------------------------------------------------
dept_counts = df["Department"].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=dept_counts.index, y=dept_counts.values)
plt.title("Patients by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 6. DIAGNOSIS FREQUENCY
# ---------------------------------------------------
diag_counts = df["Diagnosis"].value_counts().head(10)

plt.figure(figsize=(12,5))
sns.barplot(x=diag_counts.index, y=diag_counts.values)
plt.title("Top Diagnoses")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 7. TREATMENT COST BY TREATMENT TYPE
# ---------------------------------------------------
plt.figure(figsize=(10,5))
sns.barplot(
    data=df,
    x="Treatment_Type",
    y="Treatment_Cost",
    estimator="mean"
)
plt.title("Average Treatment Cost by Treatment Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 8. LENGTH OF STAY BY OUTCOME
# ---------------------------------------------------
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x="Outcome",
    y="Length_of_Stay"
)
plt.title("Length of Stay by Outcome")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 9. INSURANCE DISTRIBUTION
# ---------------------------------------------------
plt.figure(figsize=(8,4))
sns.countplot(data=df, x="Insurance")
plt.title("Insurance Coverage")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------------------------------
# 10. HEALTH METRICS CORRELATION
# ---------------------------------------------------
corr_cols = [
    "Age", "Length_of_Stay", "Treatment_Cost",
    "Blood_Pressure", "Heart_Rate",
    "Cholesterol", "BMI"
]

plt.figure(figsize=(10,6))
sns.heatmap(df[corr_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Clinical Metrics Correlation")
plt.show()
```

```python
# ---------------------------------------------------
# 11. INTERACTIVE PLOTLY — COST BY DEPARTMENT
# ---------------------------------------------------
dept_cost = (
    df.groupby("Department")["Treatment_Cost"]
    .mean()
    .reset_index()
)

fig = px.bar(
    dept_cost,
    x="Department",
    y="Treatment_Cost",
    title="Average Treatment Cost by Department"
)
fig.show()
```
